```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
def sigmoid(x):
    return 1 / (1 + np.exp(-x))
```

```python
x = np.linspace(-10, 10, 100)
sig = sigmoid(x)

plt.figure(figsize=(10, 6))
plt.plot(x, sig)
plt.grid(True);
```
