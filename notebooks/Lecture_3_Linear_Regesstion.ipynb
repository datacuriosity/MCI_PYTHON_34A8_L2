```python
x = np.linspace(1, 100, 100)
y = 2*x**2 + 3 + np.random.normal(1, 10)
```

```python
plt.figure(figsize=(10, 6))
plt.scatter(x, y)
plt.plot(x, y);
```

```python
pd.DataFrame({'X': x**2, 'y': y})
```

```python
from sklearn.linear_model import LinearRegression
```

```python
model = LinearRegression().fit((x**2).reshape(-1, 1), y)
```

```python
y_hat = model.coef_ * x**2 + model.intercept_
```

```python
plt.figure(figsize=(10, 6))
plt.scatter(x, y, color='blue', alpha=0.3);
plt.scatter(x, y_hat, color='red', alpha=0.2);
```
